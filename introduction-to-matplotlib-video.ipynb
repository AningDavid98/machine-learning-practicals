## matplotlib example workflow

```python
# prepare data 
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# setup plot
fig, ax = plt.subplots(figsize=(10,10))

# plot the data
ax.plot(x, y);

# customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis");
```

## Making figures with numpy arrays


```python
# create some data
x = np.linspace(0, 10, 100)
x
```

```python
x[:10]
```

```python
# plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# using the same data to plot a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# making plot from dictionary
nut_butter_prices = {"Almond Butter" : 10,
                     "Peanut Butter" : 8,
                     "Cashew Butter" : 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="David's Nut Butter Store",
       ylabel="Price ($)");
       
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# make some data for histogram and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# subplots
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10,10))

ax1.plot(x, x/2);
ax2.scatter(x, np.sin(x));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(x);
```

## Creating plots from pandas dataframes

```python
import pandas as pd
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('$', '').str.replace(',','').astype(float).astype(int)
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("01/02/2025", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
# let's plot the total sales
car_sales.plot("Sale Date",
               "Total Sales");
```

```python
car_sales
```

```python
car_sales.plot("Odometer (KM)", "Price", kind="scatter");
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease.info()
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,10))
scatter = ax.scatter(over_50["age"], over_50["thalach"], c=over_50["target"], cmap="winter");
ax.legend(*scatter.legend_elements(), title="Target");
ax.set(title = "Heart Disease Analysis In Conjunction With Age and Thalach",
       xlabel="Age",
       ylabel="Thalach Values");
ax.axhline(over_50["thalach"].mean(), linestyle="--");
```

```python
over_50.head()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
scatter = ax.scatter(over_50["age"], over_50["chol"], c=over_50["target"], cmap="winter");
ax.legend(*scatter.legend_elements(), title="Target");
ax.set(title="Heart Disease Analysis In Conjunction With Age and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol Level");
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
over_50.head()
```

```python
fig, ((ax1,ax2)) = plt.subplots(figsize=(10,10), nrows=2, sharex=True)
scatter = ax1.scatter(over_50["age"], over_50["thalach"], c=over_50["target"], cmap="winter");
ax1.legend(*scatter.legend_elements(), title="Target");
ax1.axhline(over_50["thalach"].mean(), linestyle="--");
ax1.set(title="Heart Disease Analysis In Conjunction With Age and Thalach Values",
        ylabel="Thalach Values");

scatter = ax2.scatter(over_50["age"], over_50["chol"], cmap="winter", c=over_50["target"]);
ax2.legend(*scatter.legend_elements(), title="Taget");
ax2.axhline(over_50["chol"].mean(), linestyle="--");
ax2.set(title="Heart Disease Analysis In Conjunction With Age and CholestrolLevel",
        xlabel="Age",
        ylabel="Cholestrol Level");

fig.suptitle("Heart Disease Analysis", fontsize=20);
```
